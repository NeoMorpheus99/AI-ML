```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("file_pe_headers.csv", sep=",")
```

```python
df.head()
```

```python
#making a scatter plot
fig = px.scatter_3d(df,
                    x="SuspiciousImportFunctions",
                    y="SectionsLength",
                    z="SuspiciousNameSection",
                    color="Malware")
fig.show()
```

```python
 #extract the Features and the target labels
 y=df["Malware"]
 x=df.drop(["Name", "Malware"], axis=1).to_numpy()
```

```python
#IMporting Sckit learn clustering Module
from sklearn.cluster import KMeans
estimator =KMeans(n_clusters=len(set(y)))
estimator.fit(x)
```

```python
y_pred = estimator.predict(x)
df["pred"] = y_pred
df["pred"]= df["pred"].astype("category")
```

```python
#making a scatter plot
fig = px.scatter_3d(df,
                    x="SuspiciousImportFunctions",
                    y="SectionsLength",
                    z="SuspiciousNameSection",
                    color="pred")
fig.show()
```
