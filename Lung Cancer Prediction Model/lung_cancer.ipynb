```python
import pandas as pd
```

```python
data = pd.read_csv('/content/lung cancer survey.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.columns = data.columns.str.strip()
```

```python
data['GENDER'] = data['GENDER'].str.strip()
```

```python
data["LUNG_CANCER"] = data['LUNG_CANCER'].str.strip()
```

```python
data.replace ({2:1, 1:0}, inplace=True)
```

```python
data.head()
```

```python
data['GENDER'] = data['GENDER'].map({'M':1, 'F':0})
```

```python
data['LUNG_CANCER'] = data['LUNG_CANCER'].map({'YES':1, 'NO':0})
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
correlation_mat = data.corr()
```

```python
correlation_mat
```

```python
correlationTarget = correlation_mat['LUNG_CANCER'].drop('LUNG_CANCER').sort_values(ascending=False).reset_index()
```
